```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating a Non-Linear Dataset

```python
X = 6 * np.random.rand(100, 1) - 3  ## np.random.rand(100, 1) returns an array of 100 rows and 1 column....2D array
y = 0.5 * X**2 + 1.5 * X + 2 + np.random.randn(100, 1)
## Quadratic equation used to make the dataset
```

```python
X
```

```python
y = y.reshape(len(y))  ## Converting 2-D to 1-D
y
```

```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

### This tells that we need to use Polynomial Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
## Lets try with Linear regression 

from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
y_test_pred = linear_regression.predict(X_test)
print(y_test_pred)

from sklearn.metrics import r2_score
score = r2_score(y_test, y_test_pred)
print(score)

## Visualize the Linear regression 
plt.scatter(X_train, y_train)
plt.xlabel('X_train')
plt.ylabel('y_train')
plt.plot(X_train, linear_regression.predict(X_train))
```

### Thus Linear Regression is underfitting

```python
## Lets apply polynomial transformation 

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=True)  ## If the dataset is having x1 feature then x1^2 will also get created
                                                        ## If the dataset is having x2 feature then x2^2 will also get created
                                                        ## inplace=True means B0 feature will also get created
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

#### Initially we had only one independent feature but now we have 3 independent features


```python
## Now we apply Multiple linear regression for this data

linear_regression_poly = LinearRegression()
linear_regression_poly.fit(X_train_poly, y_train)
print('coef : length ',len(linear_regression_poly.coef_) , ':- ', linear_regression_poly.coef_)
print('intercept : ', linear_regression_poly.intercept_)
y_test_pred_poly = linear_regression_poly.predict(X_test_poly)
print(y_test_pred_poly)

score_poly = r2_score(y_test, y_test_pred_poly)
print(score_poly)
```

```python
## Visualization of the best fit curve 

plt.scatter(X_train, y_train)
plt.scatter(X_train, linear_regression_poly.predict(X_train_poly), color='r')
```

### The red dots represents the best fit curve

```python
## Prediction of new data

Xnew = [[-1]]
Xnew = poly.transform(Xnew)
prediction = linear_regression_poly.predict(Xnew)
prediction
```
