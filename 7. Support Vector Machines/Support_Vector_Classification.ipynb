```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Creating the Dataset

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, random_state=42, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y = pd.DataFrame(X)[1], hue=y)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Model Training - Linear SVC

```python
from sklearn.svm import SVC
linear_svc = SVC(kernel='linear')  ## By default the kernel is rbf and C = 1
```

```python
linear_svc.fit(X_train, y_train)
```

```python
y_pred_linear = linear_svc.predict(X_test)
y_pred_linear
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred_linear))
print(classification_report(y_test, y_pred_linear))
```

## Model Training - rbf Kernel

```python
rbf_svc = SVC(kernel='rbf')
```

```python
rbf_svc.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf_svc.predict(X_test)
y_pred_rbf
```

```python
print(confusion_matrix(y_test, y_pred_rbf))
print(classification_report(y_test, y_pred_rbf))
```

## Model Training - Polynomial Kernel 

```python
poly_svc = SVC(kernel='poly')
```

```python
poly_svc.fit(X_train, y_train)
```

```python
y_pred_poly = poly_svc.predict(X_test)
y_pred_poly
```

```python
print(confusion_matrix(y_test, y_pred_poly))
print(classification_report(y_test, y_pred_poly))
```

## Hyperparameter Tuning 

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' : ['linear', 'rbf', 'poly', 'sigmoid']
}
```

```python
svc_model = SVC() ## By default the kernel is rbf and C = 1

model = GridSearchCV(estimator=svc_model, param_grid=params, cv = 5)

model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
